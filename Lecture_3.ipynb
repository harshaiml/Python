## List in Python
---
- A built-in data type store set of values
- It can store elements of differrent type (integer, float, string, etc.)

```python
marks = [87, 64, 33, 95, 76]     # List of student marks
student = ["Karan", 85, "Delhi"] # Student details in a list

student[0] = "Harsh"             # Allowed in Python
len(student)                     # Length of the list
```

```python
marks = [87.4, 64.5, 33.2, 95.0, 76.8]  # List of student marks as float
print(marks)                            # Print the list
print(type(marks))                      # Type of first element in the list
print(marks[0])                         # First element of the list
print(marks[-1])                        # Last element of the list
print(marks[1:4])                       # Slicing the list
marks[2] = 34.5                         # Modifying an element in the list
print(marks)                            # Print the modified list
```
