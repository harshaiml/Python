## List Slicing
---
Similer to Strings Slicing

- list_name[ starting_idx : ending_idx ]  #ending idx is not include
- marks = [87, 64, 33, 95, 76]
- marks[1 : 4] is [64, 33, 95]
- marks[ : 4 ]is name as marks[0 : 4]
- marks[ 1 : ] is same as marks[ 1 : len(marks) ]
- marks[ -3 : -1 ] is [33, 95]

```python
marks = [87, 64, 33, 95, 76] # List of student marks
print(marks[1:4])            # Slicing the list to get elements from index 1 to 3
print(marks[1:])             # Slicing the list to get elements from index 1 to end
print(marks[-3:-1])          # Slicing the list to get elements from index -3 to -2
```

## List Methods
---
- list = [2, 1, 3]
- list.append(4)               #adds one element at the end [2, 1, 3, 4]
- list.sort()                  #sorts in ascending order [1, 2, 3] 
- list.sort( reverse=True)     #sorts in descending order [3, 2, 1]
- list.insert(idx, el)         #insert element at index
- list.remove(1)               #removes first occurrence of element [2, 3, 1]
- list.pop(idx)                #remove element at inx

```python
# list.append(value)      # Adds value to the end of the list
# This is mutable operation

marks = [87, 64, 33, 95, 76]  # List of student marks
marks.append(88)              # Append 88 to the list
print(marks)                  # Print the updated list
```

```python
# list.sort() Sorts the list in ascending order

marks = [87, 64, 33, 95, 76]  # List of student marks
marks.sort()                  # Sort the list
print(marks)                  # Print the sorted list
```

```python
list = [87, 64, 33, 95, 76, 85]
print(list.sort())
print(list)
```

```python
list = [87, 64, 33, 95, 76, 85]
print(list.append(42))
print(list.sort())
print(list)
```

```python
# list.sort(reverse=True) Sorts the list in descending order

marks = [87, 64, 33, 95, 76]  # List of student marks
marks.sort(reverse=True)      # Sort the list in descending order
print(marks)                  # Print the sorted list
```

```python
# list.sort Sorts the list in descending order

marks = ["Banana", "Apple", "Cherry", "Date", "Elderberry"]  # List of student marks
print(marks.sort())                                          # Sort the list in descending order
print(marks)                                                 # Print the sorted list
```

```python
# list.sort(reverse=True) Sorts the list in descending order

marks = ['a', 'g', 'e', 'c', 'f', 'b', 'h', 'd']  # List of student marks
print(marks.sort())                               # Sort the list in descending order
print(marks)                                      # Print the sorted list
```

```python
# list.reverse() Reverses the list

marks = [87, 64, 33, 95, 76]  # List of student marks
marks.reverse()               # Reverse the list
print(marks)                  # Print the reversed list
```

```python
# list.reverse() Reverses the list

marks = ['a', 'g', 'e', 'c', 'f', 'b', 'h', 'd']  # List of student marks
marks.reverse()                                   # Reverse the list
print(marks)                                      # Print the reversed list
```

```python
# list.insert(index, value) Inserts value at the specified index

list = [87, 64, 33, 95, 76]      # List
list.insert(2, 42)               # Insert 42 at index 2
print(list)                      # Print the updated list
```

```python
# list.pop(index) Removes and returns the element at the specified index

list = [2, 1, 3, 1]           # List
list.pop(2)                   # Remove and return element at index 2
print(list)                   # Print the updated list
```

## Tuples in Python
---
A built-in data type that lets us create immutable sequences of values.
- tup = (87, 64, 33, 95, 76)   #tup[0], tup[1]..
- tup[0] = 43                  #NOT allowed in Python
- tup1 = ()
- tup2 = (1,)
- tup3 = (1, 2, 3)

```python
# tuple creation and usage

tup = (1, 2, 3, 4, 5)        # Creating a tuple
print(tup)                   # Print the tuple
print(type(tup))             # Type of the tuple
```

```python
tup = (1, 2, 3, 4, 5,)        # Creating a tuple
print(tup[0])                # Accessing first element
print(tup[1])                # Accessing second element
# tup[0] = 5                 # This will raise an error because tuples are immutable
```

```python
# empty tuple creation

tup = ()                 # Creating an empty tuple
print(tup)               # Print the empty tuple
print(type(tup))         # Type of the empty tuple
```

```python
tup = (1,)               # Creating a single-element tuple and coma is compulsory in single-element tuples
print(tup)               # Print the single-element tuple
print(type(tup))         # Type of the single-element tuple
```

```python
# Slicing in tuples

tup = (1, 2, 3, 4, 5,)
print(tup[1:4])          # Slicing from index 1 to 4
print(tup[:3])           # Slicing from beginning to index 3
print(tup[2:])           # Slicing from index 2 to end
```
