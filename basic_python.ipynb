# Variable

A Variable is a name given to a memory location in a program

```python
name = "Harsh"
age = 23
price = 25.99

age2 = age 

print(name)
print(age2)

print("My name is : ", name)
print("My age is : ", age)
```

# Rules for Identifires
---
- Identifires can be combination of uppercases and lowercases letter, digits or an underscore(_) So **myVariable, variable_1 variable_for_print** all are valid python identifires.
- An identifier can not start with digit. So while **variable1** is valid, 1variable is not valid.
- We can't use special symbols like !,#,@,%,$ etc in our identifier.
- Identifire can be of any length.

```python
# Data Types in Python
name = "Harsh"
age = 23
price = 25.99
student = True
var1 = None

print(type(name))          # String
print(type(age))           # Integer
print(type(price))         # Float
print(type(student))       # Boolean
print(type(var1))          # NoneType
```
