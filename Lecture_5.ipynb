## Loops in Python
---
Loops are used to repeat instructions.
- while Loops


```python
# while True :
#     print("Hello")

# # We can write Hello Unlimeted times using while loop 
# We can run this code
```

```python
# Print Hello five times

count = 1
while count <= 5 : 
    print("Hello")
    count += 1
```

```python
# Print Hello 5 times, and Print This is my code 5 times 

count = 1
while count <= 5 : 
    print("Hello")
    count += 1

i = 1
while i <= 5:
    print("This is my code")
    i+=1
```

```python
# print count 1 to 10

i = 1
while i <= 10:
    print(i)
    i+=1
```

```python
# I want to print numbers from 1 to 5

i = 1
while i <= 5:
    print(i)
    i += 1

print("Loop ended")
```

```python
# Print Reverse number from 1 to 5

i = 5
while i >= 1:
    print(i)
    i -= 1
print("Loop ended")
```

## Break & Continue
---
- **Break** : used to terminate the loop when encountered.
- **Countinue** : terminates executinon in the current iteration & continues execution of the loop with the next iteration.

```python
i = 1
while i <= 5:
    print(i)
    if(i == 3):
        break
    i += 1
print ("End of loop")
```

```python
# Using break

nums = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 36)

x = int(input("Enter your number"))

i = 0 # initialization
while i < len(nums):
    if(nums[i] == x):
        print("Found at index", i)
        break
    else:
        print("Finding..")
    i += 1
print("End of loop")
```

```python
# Using continue

i = 0
while i <= 5:
    if(i == 3):
        i += 1
        continue
    print(i)
    i += 1
```

```python
# Using continue
# print 1 to 20 even numbers

i = 0
while i <= 20:
    if(i%2 == 0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
# Using continue
# print 1 to 20 odd numbers

i = 0
while i <= 20:
    if(i%2 != 0):
        i += 1
        continue
    print(i)
    i += 1
```

## range()
---
Range functtion returns a sequance pf numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.
- range( start?, stop, step?)

```python
# for el in range(5):
#     print(el)

# for el in range(1, 5):
#     print(el)

# for el in range(1, 5, 2):
#     print(el)
```

```python
seq = range(10)

for i in seq:
    print(i)
```

```python
for i in range(10): # range(stop)
    print(i)
```

```python
for i in range(2, 10): # range(start, stop)
    print(i)
```

```python
for i in range(2, 10, 2): # range(start, stop, step)
    print(i)
```

```python
# print 1 to 20 even numbers

for i in range(2, 20, 2):
    print(i)
```

```python
# print 1 to 10 odd numbers

for i in range(1, 20, 2):
    print(i)
```
