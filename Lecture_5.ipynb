## Loops in Python
---
Loops are used to repeat instructions.
- while Loops


```python
# while True :
#     print("Hello")

# # We can write Hello Unlimeted times using while loop 
# We can run this code
```

```python
# Print Hello five times

count = 1
while count <= 5 : 
    print("Hello")
    count += 1
```

```python
# Print Hello 5 times, and Print This is my code 5 times 

count = 1
while count <= 5 : 
    print("Hello")
    count += 1

i = 1
while i <= 5:
    print("This is my code")
    i+=1
```

```python
# print count 1 to 10

i = 1
while i <= 10:
    print(i)
    i+=1
```

```python
# I want to print numbers from 1 to 5

i = 1
while i <= 5:
    print(i)
    i += 1

print("Loop ended")
```

```python
# Print Reverse number from 1 to 5

i = 5
while i >= 1:
    print(i)
    i -= 1
print("Loop ended")
```

## Let's Practice
---
1. Print numbers from 1 to 100.
2. Print numbers from 100 to 1.
3. Print the multiplication table of a number n.
4. Print the elements of following list using a loop:
- [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
5. Search for a number x in this tuple using loop:
- (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)

```python
# 1. Print numbers from 1 to 100.

i = 1
while i <= 100:
    print(i)
    i+=1
```

```python
# 2. Print numbers from 100 to 1.

i = 100
while i >= 1: #stoping value
    print(i)
    i -= 1
```

```python
# 3. Print the multiplication table of a number n.

n = int(input("Enter your table number : "))
i = 1
while i <= 10:
    print(n*i)
    i += 1
```

```python
# 4. Print the elements of following list using a loop:
# - [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]

nums = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]

idx = 0
while idx < len(nums):
    print(nums[idx])
    idx += 1
```

```python
heroes = ["ironman", "thor", "superman", "batman", "groot"]

idx = 0
while idx < len(heroes):
    print(heroes[idx])
    idx += 1
```

```python
# 5. Search for a number x in this tuple using loop:
# - (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)

nums = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100, 36)

x = int(input("Enter your number"))

i = 0 # initialization
while i < len(nums):
    if(nums[i] == x):
        print("Found at index", i)
    else:
        print("Finding..")
    i += 1
```
