## Loops in Python
---
Loops are used to repeat instructions.
- while Loops


```python
# while True :
#     print("Hello")

# # We can write Hello Unlimeted times using while loop 
# We can run this code
```

```python
# Print Hello five times

count = 1
while count <= 5 : 
    print("Hello")
    count += 1
```

```python
# Print Hello 5 times, and Print This is my code 5 times 

count = 1
while count <= 5 : 
    print("Hello")
    count += 1

i = 1
while i <= 5:
    print("This is my code")
    i+=1
```

```python
# print count 1 to 10

i = 1
while i <= 10:
    print(i)
    i+=1
```

```python
# I want to print numbers from 1 to 5

i = 1
while i <= 5:
    print(i)
    i += 1

print("Loop ended")
```

```python
# Print Reverse number from 1 to 5

i = 5
while i >= 1:
    print(i)
    i -= 1
print("Loop ended")
```
