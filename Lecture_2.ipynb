## String Function
---
- str = "I am a coder."
- str.endswith("er.") #return true if string ends with substr
- str.capitalize() #capitalize 1st char
- str.replace(old,new) #replaces all occurrences of old with new
- str.find(word) #return 1st index of 1st occurrence
- str.count("am") #counts the occurrence of substr in string

```python
# str.endswith(suffix)

str = "learning Python from Vishlesan i-HUB IIT Patna"
print(str.endswith("tna")) # True
```

```python
# str.endswith(suffix)

str = "learning Python from Vishlesan i-HUB IIT Patna"
print(str.endswith("hff")) # False
```

```python
# str.capitalize()

str = "learning Python from Vishlesan i-HUB IIT Patna"
print(str.capitalize()) # Learning python python from vishlesan i-hub iit patna
```

```python
# str.capitalize()

str = "learning Python from Vishlesan i-HUB IIT Patna"
str = str.capitalize()
print(str)
```

```python
# str.replace(old, new)

str = "Learning Python from Vishlesan i-HUB IIT Patna"
print(str.replace("Python", "AI&ML"))
```

```python
# str.find(substring)

str = "Learning Python from Vishlesan i-HUB IIT Patna"
print(str.find("Vishlesan"))  # 21
```

```python
# str.count(substring)

str = "Learning Python from Vishlesan i-HUB IIT Patna"
print(str.count("Python"))  # 1
```

## Conditional Statements
---

```python
# if-elif-else(SYNTAX)
    
#     if(condition):

#         Statement1

#     elif(condition):

#         Statement2

#     else:

#         StatementN
```

```python
# if condition:

age = int(input("Enter your age:"))

if(age >= 18):
    print("You are eligible for driving license.")
    
```

```python
# elif condition:

light = "green"

if (light == "red"):
    print ("Stop the vehicle.")
elif (light == "yellow"):
    print ("Get ready to move.")
elif (light == "green"):
    print ("You can move now.")
print("Enjoy your ride!")
```

```python
num = 5

if(num > 2):
    print("Greater than 2")
elif(num > 3):
    print("Greater than 3")
```

## Conditional Statements
---
- marks >= 99,grade = "A"
- 90 > marks >= 80, grade = "B"
- 80 > marks >= 70, grade = "C"
- 70 > marks, grade = "D"

```python
marks = float(input("Enter Student marks: "))

if(marks >= 90):
    grade = "A"
elif(marks >= 80 and marks < 90):
    grade = "B"
elif(marks >= 70 and marks < 80):
    grade = "C"
else:
    grade = "D"

    print("Grade of the student ->", grade)
```

## Nesting
---

```python
# Nested if-else

age = 95

if(age >= 18):
    if(age >= 80):
        print("You need a senior citizen driving license.")
    print("You can drive")
else:
    print("You cannnot drive")
```

```python
# 1. WAP to check if a number entered by the user is odd or even.

num = int(input("Enter a number:"))

rem = num % 2

if(rem == 0):
    print("Even Number")
else:
    print("Odd Number")
```

```python
# 2. WAP to find the greatest of 3 numbers entered by the user.

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

if(num1 >= num2 and num1 >= num3):
    print("First number is largest", num1)
elif(num2 >= num3):
    print("Second number is largest", num2)
else:
    print("Third number is largest", num3)
```
