```python
# Character                        Meaning
    
#    'r'         open for reading (default)
#    'w'         open for writing, truncating the file first
#    'x'         create a new file and open if for writing
#    'a'         open for writing, appending to the end of the file if it exists
#    'b'         binary mode
#    't'         text mode (default)
#    '+'         open a disk file for updating (reading and writing)
```

```python
# 1. 1. Create a new file "practice.text" using Python. And the following datain it:
# - Hi everyone
# - we are learning File I/O
# - using java
# - i like programing in Python


with open("practice.txt", "w") as f:
    f.write("Hi everyone\n")
    f.write("we are learning File I/O\n")
    f.write("using Java\n")
    f.write("i like programing in Python\n")
```

```python
# 2. WAF that replace all occurrences of "java" with "Python" in above file.


with open("practice.txt", "r") as f:
    data = f.read()

new_data = data.replace("Java", "Python")
print(new_data)

with open("practice.txt", "w") as f:
    data = f.write(new_data)
```

```python
# 3.Search if the word "learning" exists in the file or not.

def check_for_word():
    word = "learning"
with open("practice.txt", "r") as f:
    data = f.read()
    if(data.find(word) != -1):
        print("Word found")
    else:
        print("Word not found")
check_for_word()
```

```python
# 4. WAF to find in which line of the file does the word "Python" occur first.
# - Print-1 if word not found.

def check_word_line():
    word = "Python"
    data = True
    line_no = 1
    with open("practice.txt", "r") as f:
        while data:
            data = f.readline()
            if(word in data):
                print(line_no)
                return
            line_no += 1
        
    return -1
check_word_line()
```

```python
# 5. From a file containing numbers sparated by comma, print the cound of even numbers.

with open("numbers.txt", "r") as f:
    data = f.read()
    print(data)

    num = ""
    for i in range(len(data)):
        if data[i] == ",":
            print(int(num))
            num = ""
        else:
            num += data[i]
```

```python
# 5. From a file containing numbers sparated by comma, print the cound of even numbers.
 
def count_even_numbers():
    count = 0
    with open("numbers.txt", "r") as f:
        data = f.read()
        print(data)

        numbers = data.split(",")
        for val in numbers:
            if(int(val) % 2 == 0):
                count += 1

    print("Count of even numbers:", count)

count_even_numbers()
```
