# Rules for Identifires
---
- Identifires can be combination of uppercases and lowercases letter, digits or an underscore(_) So **myVariable, variable_1 variable_for_print** all are valid python identifires.
- An identifier can not start with digit. So while **variable1** is valid, 1variable is not valid.
- We can't use special symbols like !,#,@,%,$ etc in our identifier.
- Identifire can be of any length.

## Types of Operators
---
An operation is a symbol that performs a certain between operands.

- Arithmetic Operators (+,-,*,/,%,**)
- Relation / Comparison Operators (==,!=,>,<,>=,<=)
- Assignment Operators (=, +=, -=, *=, /=, %=, **= )
- Logical Operators (not, and, or)

```python
# Arithmetic Operations

a = 15
b = 4

print("Addition:", a + b)            # Addition
print("Subtraction:", a - b)         # Subtraction
print("Multiplication:", a * b)      # Multiplication
print("Division:", a / b)            # Float division
print("Modulus:", a % b)             # Remainder
print("Exponentiation:", a ** b)     # Power
```

```python
# Relational Operators

x = 50
y = 20

print("Greater than:", x > y)               # Greater than
print("Less than:", x < y)                  # Less than
print("Equal to:", x == y)                  # Equal to
print("Not equal to:", x != y)              # Not equal to
print("Greater than or equal to:", x >= y)  # Greater than or equal to
print("Less than or equal to:", x <= y)     # Less than or equal to
```

## input()
---
In Python, input() is a built-in function used to receive data from the user via the keyboard while a program is running.

- input() statement is used to accept values (using keyboard) from user
- input() # result for input() is always str
- int ( input()) #int
- float ( input()) #float

```python
name = input("Enter your name: ")
print("Hello", name)
```

```python
age = input("Enter your age: ")
print("You are", age, "years old.")
```

```python
val = input("Enter some value: ")
print(type(val), val)
```

```python
# Type casting input

val = input("Enter some value: ")
print(type(val), val)
```

```python
# Type casting input to int

val = int( input("Enter some integer value: ")) #int
print(type(val), val)
```

```python
# Type casting input to float

val = float( input("Enter some float value: ")) #float
print(type(val), val)
```

```python
name = input("Enter your name: ")
age = int( input("Enter your age: "))
marks = float( input("Enter your marks: "))

print("Welcome ", name)
print("Age :", age)
print("Marks :", marks)
```
