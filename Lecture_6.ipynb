---

## Let's Practice
---
1. WAF to print the lenght of a list. (list is the parameter)
2. WAF to print the elements of a list in a single line. (list is the parameter)
3. WAF to find the factorial of no.(n is the parameter)
4. WAF to convert USD to INR.

```python
# 1. WAF to print the lenght of a list. (list is the parameter)

cities = ["Delhi", "Gurgaon", "Noida", "Pune", "Mumbai", "Chennai"]
heroes = ["Spider-Man", "Captin America", "Iron Man", "Thor", "Hulk", "Wolverine", "Black Panther", "Captian Marvel", "Docter Strange", "Scarlet Witch"]

def print_len(list):
    print(len(list))

print_len(cities)
print_len(heroes)
```

```python
# 2. WAF to print the elements of a list in a single line. (list is the parameter)

cities = ["Delhi", "Gurgaon", "Noida", "Pune", "Mumbai", "Chennai"]
heroes = ["Spider-Man", "Captin America", "Iron Man", "Thor", "Hulk", "Wolverine", "Black Panther", "Captian Marvel", "Docter Strange", "Scarlet Witch"]

def print_len(list):
    print(len(list))

def print_list(list):
    for item in list:
        print(item, end=" ")

print_list(heroes)
print_list(cities)
```

```python
# 3. WAF to find the factorial of no.(n is the parameter)

def cal_fact(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    print(fact)

cal_fact(6)
```

```python
# 4. WAF to convert USD to INR.

def converter(usd_val,):
    inr_val = usd_val * 90
    print(usd_val, "USD =", inr_val, "INR")

converter(100)
```

## Recursion
---
When a function calls itself repeatedly.

```python
# #print n to 1 backwards

# def show(n):
#     if (n == 0):
#         return
#     print(n)
#     show(n-1)
```

```python
def show(n):
    print(n)

show(5)  #5, 4, 3, 2, 1
```

```python
#recursive function

def show(n):
    if(n == 0):
        return
    print(n)
    show(n-1)
    print("End")

show(5)  #5, 4 = n-1, 3 = n-2, 2 = n-3, 1
```
