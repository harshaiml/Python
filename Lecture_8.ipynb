## OOP in Python
---
- To map with real world scenarios, we started usingobjects in code.
- This is called object oriented programming

```python
class Student:           # creating a class
    name = "Karan"       # creating a class variable

s1 = Student()           # Creating an object of the class
print(s1.name)           # Output : Karan

s2 = Student()           # Creating another object of the class
print(s2.name)           # Output : Karan
```

```python
class Car:
    color = "Black"
    brand = "Ford"

c1 = Car()
print(c1.color)     # Output : Black
print(c1.brand)     # Output : Ford
```

```python
class Student:           # creating a class
    name = "Karan"       # creating a class variable
    def __init__(self):
        print(self)
        print("Adding new student in Database.")


s1 = Student()
print(s1)
```

```python
class Student:      
    
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        print("Adding new student in Database.")


s1 = Student("Harsh", 95)
print(s1.name, s1.marks) # Harsh 95

s2 = Student("Anjali", 85)
print(s2.name, s2.marks) # Anjali 85
```

```python
class Student:

    #default constructor
    def __init__(self):
        pass

    # Prameterized constructor
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        print("Adding new student in Database.")


s1 = Student("Harsh", 95)
print(s1.name, s1.marks) # Harsh 95

s2 = Student("Anjali", 85)
print(s2.name, s2.marks) # Anjali 85
```

## Class & Instance Attributes
---
- Class.attr
- obj.attr

```python
class Student:

    college_name = "ABC College" 

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        print("Adding new student in Database.")


s1 = Student("Harsh", 95)
print(s1.name, s1.marks) # Harsh 95

s2 = Student("Anjali", 85)
print(s2.name, s2.marks) # Anjali 85

print(s2.college_name)
```

```python
class Student:

    college_name = "ABC College" 
    name = "Anonymous" # if any student name is not provided, this will be the default name

    def __init__(self, name, marks):
        self.name = name      # Always give the higher priority to instance variable
        self.marks = marks
        print("Adding new student in Database.")


s1 = Student("Harsh", 95)
print(s1.name, s1.marks)
```
