## Let's Practice
---
1. Create student class that taken name & mark of 3 subjects as arguments in constructor.
- Then create a method to print avrage.

```python
# 1. Create student class that taken name & mark of 3 subjects as arguments in constructor.
# - Then create a method to print avrage.

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_avrage(self):
        sum = 0
        for val in self.marks:
            sum += val
        print("Hii", self.name, "Your Avrage score is:", sum / len(self.marks))


s1 = Student("Alice", [99, 98, 97])
s1.get_avrage()

s1.name = "Bob"
s1.get_avrage()
```

## Important
---
1. Abstraction
- Hiding the implementntaion details of a class and only showing the essential features to the user.
2. Encapsulation 
- Wrapping data and functions into a single unit (object).

```python
# Abstraction Example

class Car:
    def __init__(self):
        self.accelerate = False
        self.breakes = False
        self.clutch = False

    def start_car(self):
        self.accelerate = True
        self.breakes = True
        self.clutch = True
        print("Car started.")

car1 = Car()
car1.start_car()
```

```python
# 1. Create Account class with 2 attributes - balance & account no.
# - Create methods for debit, credit & printing the blance.

class Account:
    def __init__(self, blance, account_no):
        self.blance = blance
        self.account = account_no


    #debit method
    def debit(self, amount):
        self.blance -= amount
        print("Rs.", amount, "Debited in your account.")
        print("Your current blance is Rs.", self.blance)

    #credit method
    def credit(self, amount):
        self.blance += amount
        print("Rs.", amount, "Credited in your account.")
        print("Your current blance is Rs.", self.blance)

    def get_blance(self):
        return self.blance


account1 = Account(10000, "ACC.XXXXXXXXXX126")
print(account1.blance)
print(account1.account)

account1.debit(2000)
account1.credit(8658)
account1.credit(6000)
account1.credit(8456)
account1.credit(15646)
account1.credit(8564)

```
