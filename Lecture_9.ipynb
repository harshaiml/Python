## del keyword
---
Used to delete object properties or itself.
- del s1.name
- del s1

```python
# class Student:
#     def __init__(self, name):
#         self.name = name

#     s1 = Student("Alice")

#     del s1
#     print(s1.name) # This will raise a NameError since s1 has been deleted
```

```python
class Student:
    def __init__(self, name):
        self.name = name

s1 = Student("Alice")
print(s1.name)  # This will print "Alice"

del s1
print(s1.name)  # This will raise a NameError since s1 has been deleted
```
