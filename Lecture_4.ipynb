## Dictionary in Python
---
- Dictionaries are used to store data value in **key:value** pairs
- They are unordered, mutable(changeable) & don't allow duplicate keys
- dict["name"],dict["cgpa"], dict["marks"]
- dict["key"] = "value" #to assign or add new

```python
# Creating dictionary

info = {                         # Dictionary starts here
    "key" : "value",             # key-value pair
    "name" : "Alice",            # Another key-value pair
    "learning" : "Python",       # Another key-value pair
    "age" : 25,                  # Integer value
    "is_adult" : True,           # Boolean value
    "marks" :85.5                # Float value
}                                # Dictionary ends here

print(info)                      # Print the entire dictionary
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

print(info)                                               # Print the entire dictionary
print(type(info))                                         # Print the type of 'info' to confirm it's a dictionary
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

print(info["name"])                                       # Accessing value using key "name"
print(info["subject"])                                    # Accessing value using key "subject"
print(info["topics"])                                     # Accessing value using key "topics"
print(info["age"])                                        # Accessing value using key "age"
print(info["is_adult"])                                   # Accessing value using key "is_adult"
print(info[12.99])                                        # Accessing value using float key 12.99
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

info["name"] = "Bob"                                      # Modifying value for key "name"
info["surname"] = "Smith"                                 # Adding new key-value pair
print(info)                                               # Print the entire dictionary
```

```python
# null dictionary

info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

null_dict = {}                                            # Creating an empty dictionary
null_dict["name"] = "Alice"                             
print(null_dict)
```

## Dictionary in Python

**Nested** Dictionaries

```python
# Nested Dictionaries

student = {
    "name" :"Shiv",
    "subjects" : {
        "phy" : 97,
        "chem" : 98,
        "math" : 95
    }
}

# This is Nested Dictionary

print(student)
print(student["subjects"])
print(student["subjects"]["chem"])
```

## Set in Python
---
1. Set in the collection of the unordered items.
2. Each element in the set must be unique & immuteble.
- nums = {1, 2, 3, 4}
- set2 = {1, 2, 2, 2}
- #repeated elements stored only once, so it resolved to {1, 2}
- null_set = set() #empty set syntax

```python
collection = {1, 2, 3, 4}

print(collection)
print(type(collection))
```

```python
# set

collection = {1, 2, 2, 2, "hello", "world", "world", 4}

print(collection)
print(type(collection))
print("Lenght of collection: ",len(collection)) # Duplicate values are not allowed in sets
```

```python
# empty set creating

collection = set()  #empty seet: syntax

print(collection)
print(type(collection))
```

## Set Methods
---
- set.add(element) # add an element
- set.remove(element) # removes the elem an
- set.clear() # empties the set
- set.pop() # removes a random value
- set.union(set2) # combines both set values & return new
- set.intersection(set2) # combines common value & return new

```python
# set.add(element)

collection = set()  # empty seet: syntax
collection.add(1)   # add 1st element
collection.add(2)   # add 2nd element
collection.add(3)   # add 3rd element
collection.add(4)   # add 4th element

print(collection)   # print set
```

```python
# set.remove(element)

collection = set()       # empty seet: syntax
collection.add(1)        # add 1st element
collection.add(2)        # add 2nd element
collection.add(3)        # add 3rd element
collection.add(4)        # add 4th element

collection.remove(4)     # removing 4th element
print(collection)        # print set
```

```python
# set.remove(element)

collection = set()          # empty seet: syntax
collection.add(1)           # add 1st element
collection.add("Alice")     # add string element
collection.add((1, 2, 3))   # add tuple element
#collection.add([1, 2, 3])  # we can't add list element it will give error

print(collection)           # print set
```

```python
# set.clear

collection = set()          # empty seet: syntax
collection.add(1)           # add 1st element
collection.add(2)           # add 2nd element
collection.add("Alice")     # add string element
collection.add((1, 2, 3))   # add tuple element

collection.clear()          # clear set
print(collection)           # print set
```
