## Dictionary in Python
---
- Dictionaries are used to store data value in **key:value** pairs
- They are unordered, mutable(changeable) & don't allow duplicate keys
- dict["name"],dict["cgpa"], dict["marks"]
- dict["key"] = "value" #to assign or add new

```python
# Creating dictionary

info = {                         # Dictionary starts here
    "key" : "value",             # key-value pair
    "name" : "Alice",            # Another key-value pair
    "learning" : "Python",       # Another key-value pair
    "age" : 25,                  # Integer value
    "is_adult" : True,           # Boolean value
    "marks" :85.5                # Float value
}                                # Dictionary ends here

print(info)                      # Print the entire dictionary
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

print(info)                                               # Print the entire dictionary
print(type(info))                                         # Print the type of 'info' to confirm it's a dictionary
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

print(info["name"])                                       # Accessing value using key "name"
print(info["subject"])                                    # Accessing value using key "subject"
print(info["topics"])                                     # Accessing value using key "topics"
print(info["age"])                                        # Accessing value using key "age"
print(info["is_adult"])                                   # Accessing value using key "is_adult"
print(info[12.99])                                        # Accessing value using float key 12.99
```

```python
info = {                                                  # Dictionary starts here
    "name" : "Alice",                                     # Another key-value pair
    "subject" : ["Python", "Java", "C++"],                # List as a value
    "topics" : ("OOP", "Data Structures", "Algorithms"),  # Tuple as a value
    "age" : 25,                                           # Integer value
    "is_adult" : True,                                    # Boolean value
    12.99 : 85.2                                          # Float as a key
}                                                         # Dictionary ends here

info["name"] = "Bob"                                      # Modifying value for key "name"
info["surname"] = "Smith"                                 # Adding new key-value pair
print(info)                                               # Print the entire dictionary
```
